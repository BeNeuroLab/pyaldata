## Inspect Field Types and Shapes
For each column, show the dtype, Python type, and shape (if array) of the first entry.

```python
rows = []
for col in df.columns:
    val = df[col].iloc[0]
    type_name = type(val).__name__
    if isinstance(val, np.ndarray):
        shape_str = str(val.shape)
        dtype_str = str(val.dtype)
    else:
        shape_str = "scalar"
        dtype_str = type_name
    rows.append({"column": col, "python_type": type_name, "dtype": dtype_str, "shape_row0": shape_str, "example": repr(val)[:80]})

field_info = pd.DataFrame(rows)
pd.set_option("display.max_rows", None)
pd.set_option("display.max_colwidth", 80)
field_info
```

## Identify Time-Varying vs Scalar Fields

```python
try:
    tv_fields = pyaldata.get_time_varying_fields(df)
    print("Time-varying fields:")
    for f in tv_fields:
        val = df[f].iloc[0]
        shape = val.shape if isinstance(val, np.ndarray) else "N/A"
        print(f"  {f:30s} shape={shape}")
except Exception as e:
    print(f"Could not auto-detect time-varying fields: {e}")
    # Fallback: find array fields and check lengths
    arr_fields = pyaldata.get_array_fields(df)
    print(f"\nArray fields: {arr_fields}")

print("\n--- Scalar / metadata fields ---")
for col in df.columns:
    val = df[col].iloc[0]
    if not isinstance(val, np.ndarray):
        print(f"  {col:30s} type={type(val).__name__:10s} example={repr(val)[:60]}")
```

## Inspect All Public pyaldata Functions

```python
public_funcs = [name for name in sorted(dir(pyaldata)) if not name.startswith('_') and callable(getattr(pyaldata, name))]
print(f"Total public functions: {len(public_funcs)}\n")

for name in public_funcs:
    fn = getattr(pyaldata, name)
    try:
        sig = inspect.signature(fn)
    except (ValueError, TypeError):
        sig = "(N/A)"
    doc = (fn.__doc__ or "No docstring").strip().split('\n')[0]
    print(f"  {name}{sig}")
    print(f"    → {doc}\n")
```

## Correctness & Performance Comparison: Original vs Refactored

We run every testable pyaldata function with **both** the original code (via `git stash`) and the refactored code, then compare:
1. **Correctness** — outputs must match exactly (or within float tolerance)
2. **Speed** — wall-clock timing comparison

```python
## Step 1: Generate baseline results with ORIGINAL (pre-refactor) code
import subprocess, os, pickle

os.chdir(r'c:\repos\pyaldata')

# Stash refactored changes → original code is on disk
stash_result = subprocess.run(['git', 'stash'], capture_output=True, text=True)
print("git stash:", stash_result.stdout.strip())

try:
    # Run baseline_test.py in a fresh Python process (sees original code)
    result = subprocess.run(
        ['python', 'baseline_test.py'],
        capture_output=True, text=True,
        cwd=r'c:\repos\pyaldata',
        timeout=600
    )
    print(result.stdout)
    if result.returncode != 0:
        print("ERRORS:", result.stderr[-3000:])
finally:
    # Always restore refactored code
    pop_result = subprocess.run(['git', 'stash', 'pop'], capture_output=True, text=True)
    print("git stash pop:", pop_result.stdout.strip())

# Load baseline
baseline = pickle.load(open(r'c:\repos\pyaldata\baseline_results.pkl', 'rb'))
print(f"\n✓ Baseline loaded: {len(baseline['timings'])} functions timed")
```

```python
## Step 2: Run the SAME tests with REFACTORED code (already in kernel)
import importlib, time, warnings
warnings.filterwarnings('ignore')

# Clean-reimport all pyaldata modules from the (restored) refactored files on disk
for mod_name in list(sys.modules.keys()):
    if mod_name.startswith('pyaldata'):
        del sys.modules[mod_name]
if 'baseline_test' in sys.modules:
    del sys.modules['baseline_test']

import pyaldata
import numpy as np

# Re-load the data with refactored mat2dataframe
mat_path = r"C:\data\raw\M079\M079_2025_09_26_10_30\M079_2025_09_26_10_30_pyaldata.mat"
t0 = time.perf_counter()
for _ in range(3):
    df = pyaldata.mat2dataframe(mat_path, shift_idx_fields=False)
new_load_time = (time.perf_counter() - t0) / 3
print(f"mat2dataframe: {new_load_time*1000:.1f} ms  (shape={df.shape})")

# Import and run the shared test function (it will import the fresh refactored pyaldata)
from baseline_test import run_all_tests

new_results, new_timings = run_all_tests(df)
new_timings['mat2dataframe'] = new_load_time
print(f"\n✓ Refactored tests done: {len(new_timings)} functions timed")
```

```python
## Step 3: Compare outputs — every result must match
bl_results = baseline['results']
bl_timings = baseline['timings']

def compare_values(old, new):
    """Compare two test outputs, return True if they match."""
    if isinstance(old, np.ndarray) and isinstance(new, np.ndarray):
        if old.shape != new.shape:
            return False
        if old.dtype.kind == 'f' or new.dtype.kind == 'f':
            return np.allclose(old, new, equal_nan=True, rtol=1e-6)
        return np.array_equal(old, new)
    elif isinstance(old, (list, tuple)) and isinstance(new, (list, tuple)):
        if len(old) != len(new):
            return False
        return all(compare_values(o, n) for o, n in zip(old, new))
    elif isinstance(old, slice) and isinstance(new, slice):
        return old.start == new.start and old.stop == new.stop
    elif isinstance(old, float) and isinstance(new, float):
        return np.isclose(old, new, equal_nan=True)
    else:
        return old == new

# Check every result key
all_keys = sorted(set(bl_results.keys()) & set(new_results.keys()))
matches = {}
for key in all_keys:
    try:
        matches[key] = compare_values(bl_results[key], new_results[key])
    except Exception as e:
        matches[key] = False
        print(f"  ⚠ Error comparing {key}: {e}")

n_pass = sum(matches.values())
n_total = len(matches)
print(f"Correctness: {n_pass}/{n_total} tests match\n")
for key, ok in sorted(matches.items()):
    status = "✓" if ok else "✗ MISMATCH"
    print(f"  {status}  {key}")

if n_pass == n_total:
    print(f"\n✓ ALL {n_total} outputs are identical — no functional changes.")
else:
    print(f"\n✗ {n_total - n_pass} mismatches found!")
```

```python
## Step 4: Speed comparison table
import pandas as pd

common_funcs = sorted(set(bl_timings.keys()) & set(new_timings.keys()))

rows = []
for fn in common_funcs:
    old_ms = bl_timings[fn] * 1000
    new_ms = new_timings[fn] * 1000
    speedup = old_ms / new_ms if new_ms > 0 else float('inf')
    rows.append({
        'function': fn,
        'original (ms)': round(old_ms, 3),
        'refactored (ms)': round(new_ms, 3),
        'speedup': round(speedup, 2),
    })

timing_df = pd.DataFrame(rows).sort_values('speedup', ascending=False)
timing_df = timing_df.reset_index(drop=True)

# Highlight functions that got faster
pd.set_option('display.max_rows', None)
pd.set_option('display.float_format', '{:.3f}'.format)

print("Speed comparison (sorted by speedup, descending):\n")
print(timing_df.to_string(index=False))

# Summary stats
faster = timing_df[timing_df['speedup'] > 1.05]
slower = timing_df[timing_df['speedup'] < 0.95]
same = timing_df[(timing_df['speedup'] >= 0.95) & (timing_df['speedup'] <= 1.05)]
print(f"\n── Summary ──")
print(f"  Faster (>5%):    {len(faster)} functions")
print(f"  Same (±5%):      {len(same)} functions")
print(f"  Slower (<-5%):   {len(slower)} functions")
if len(faster) > 0:
    print(f"  Max speedup:     {faster['speedup'].max():.1f}x  ({faster.iloc[0]['function']})")
    total_old = timing_df['original (ms)'].sum()
    total_new = timing_df['refactored (ms)'].sum()
    print(f"  Total wall time: {total_old:.0f} ms → {total_new:.0f} ms  ({total_old/total_new:.2f}x overall)")
```
